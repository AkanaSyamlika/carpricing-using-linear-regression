```python
data.drop(['lat','ID'],axis=1)#drop is used delete
```

```python
data1=data.drop(['lat','ID'],axis=1)
```

```python
data2=data1.drop('lon',axis=1)
```

```python
display(data2)
```

```python
data3=a.get_dummies(data2,dtype=int) #dummies is used to
data3
```

```python
y=data3['price']
x=data3.drop('price',axis=1)
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_test.head(5)
```

```python
x_train.shape
```

```python
y_train
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)

```

```python
ypred=reg.predict(x_test)
```

```python
ypred
```

```python
from sklearn.metrics import r2_score#metrics used to predict
r2_score(y_test,ypred)
```

```python

from sklearn.metrics import mean_squared_error
mean_squared_error(ypred,y_test)
```

```python
Results=a.DataFrame(columns=['Price','Predicted'])
Results['Price']=y_test
Results['Predicted']=ypred
Results=Results.reset_index()
Results['Id']=Results.index
Results.head(15)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lineplot(x='Id',y='Price',data=Results.head(50))
sns.lineplot(x='Id',y='Predicted',data=Results.head(50))
plt.plot()
```
